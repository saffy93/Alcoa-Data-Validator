# ALCOA+ Data Validator



**A simple Python tool to validate data integrity using ALCOA+ principles**



This project demonstrates practical application of ALCOA+ (Attributable, Legible, Contemporaneous, Original, Accurate, Complete, Consistent, Enduring, Available) — core principles used daily in regulated histology laboratories and clinical environments.

```python
def alcoa_validator(record):

    """

    Validates a data record against ALCOA+ principles.

    Returns '✅ PASS' if all principles are satisfied, otherwise returns failure reasons.

    """

    issues = []



    if not record.get('attributable'):

        issues.append("Missing Attributable (who performed the task)")

    if not record.get('legible'):

        issues.append("Not Legible")

    if not record.get('contemporaneous'):

        issues.append("Not Contemporaneous (recorded too late)")

    if not record.get('original'):

        issues.append("Not Original")

    if not record.get('accurate'):

        issues.append("Not Accurate")

    if not record.get('complete'):

        issues.append("Not Complete")

    if not record.get('consistent'):

        issues.append("Not Consistent")

    if not record.get('enduring'):

        issues.append("Not Enduring")

    if not record.get('available'):

        issues.append("Not Available")



    if not issues:

        return "✅ PASS"

    else:

        return f"❌ FAIL: {', '.join(issues)}"





# Test records (realistic examples from lab work)

records = [

    {

        'attributable': True,

        'legible': True,

        'contemporaneous': True,

        'original': True,

        'accurate': True,

        'complete': True,

        'consistent': True,

        'enduring': True,

        'available': True

    },

    {

        'attributable': False,   # missing who did it

        'legible': True,

        'contemporaneous': True,

        'original': True,

        'accurate': True,

        'complete': True,

        'consistent': True,

        'enduring': True,

        'available': True

    },

    {

        'attributable': True,

        'legible': True,

        'contemporaneous': False,  # recorded too late

        'original': True,

        'accurate': True,

        'complete': True,

        'consistent': True,

        'enduring': True,

        'available': True

    }

]



# Run the validator on each record

print("=== ALCOA+ Data Validator Test Results ===\n")

for i, record in enumerate(records, 1):

    result = alcoa_validator(record)

    print(f"Record {i}: {result}")
```
